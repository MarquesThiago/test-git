```python
# facade 
class ManagmentEvent:
    
    def __init__(self):
        print("start prepratives from event")
        
    def organize(self):
        self.space = Space()
        self.space.reserve()
        
        self.foot = Foot()
        self.foot.hire()
        
        self.music = HouseSound()
        self.music.hire_dj()
        
        self.decoration = Decorator()
        self.decoration.but_flowers()
```

```python
# subsistem 1
class Space:
    
    def __init__(self):
        self._avaliable = True
        print("hire space to your event")
        
        
    def reserve(self):
        if self._avaliable:
            print("accept: space reserved")
        else:
            print("space in used, you can't reserve")
            
# subsistem 2            
class Foot:
    
    def __init__(self):
        print("hire buffet to your event")
        
        
    def hire(self):
        print("accept: hire buffet ")
        
# subsistem 3
class HouseSound:
    
    def __init__(self):
        print("Hire band or DJ to your event")
        
        
    def hire_dj(self):
        print("accept: DJ accept your proposT")
        
# subsistem 4
class Decorator:
    
    def __init__(self):
        print("Hire decations to event")
        
        
    def but_flowers(self):
        print("but flowers")


```

```python
class Client:
    
    def __init__(self):
        print("I need organize my event wedding")
    
    def hire_professional(self):
        managment = ManagmentEvent()
        managment.organize()
        
    # function delete object    
    def __del__(self):
        print("Thank a lot, good bye")
```

```python
cli = Client()
```
