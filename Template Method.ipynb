# Template Method

O metodo template permite que apartir de uma classe abstrata permite padronizar a execução dos metodos de uma classe entretando disponibiliza os metodos como metodos abstratos para as classes que o implementem da forma que lhe for conveniente 

```python
from abc import ABCMeta, abstractmethod
```

```python
class PrepareDrinkHot(metaclass= ABCMeta):
    
    @abstractmethod
    def boil(self):
        pass
    
    @abstractmethod
    def prepare(self):
        pass
    
    @abstractmethod
    def sleep(self):
        pass
    
    def make(self):
        self.boil()
        self.prepare()
        self.sleep()
```

```python
class Koohii(PrepareDrinkHot):
    
    def boil(self):
        print("fever agua á 100ªC por 15")
    
    def prepare(self):
        print("colocando pó no coador")
        print("jogando agua")
        print("colocar na caneca")
    
    def sleep(self):
        pass
```

```python
class Ocha(PrepareDrinkHot):
    
    def boil(self):
        print("fever agua á 100ªC por 15")
    
    def prepare(self):
        print("colocando pó na xicara")
        print("jogando agua quente")
    
    def sleep(self):
        print("esperar 2 minutos")
```

```python
cafe = Koohii()
cha = Ocha()
```

```python
cha.make()
```

```python
cafe.make()
```

# vantagens

* Uma das principais vantagens é que não há duplicação de código;

* A reutilização de código ocorre normalmente já que o padrão Template Method utiliza
herança e não composição. Somente alguns métodos precisam ser sobrescritos.

* A flexibilidade permite que as subclasses decidam como implementar os passos do algoritmo

# desvantagens

* Debugar a aplicação e compreender a sequência do fluxo no padrão Template Method às
vezes pode ser confuso. Você pode acabar implementando um método que não deveria ser
implementado ou deixando de implementar um método abstrato. Vale a pena documentar
bem o código e realizar tratamento de erros para evitar problemas;

* A manutenção dos componentes de alto ou baixo nível pode causar problemas na
implementação;

## Principio de Hollywood

O Princípio de Hollywood é um princpipio de design de software que diz:

Não ligue para nós; nós ligaremos para você.
Isto é provenidente da filosofia de Hollywood, segundo a qual as produtoras de filmes e séries
ligam para os atores caso haja algum papel para eles.

#### no template method

Isso se relaciona ao padrão Template Method no sentido de que é a classe abstrata de alto nível
que organiza os passos para definir o algoritmo.

Com base no algoritmo, as classes de baixo nível são solicitadas a definir a implementação
concreta dos passos.
