# Proxy 

Cria um camada entre o objeto com exatamente as funcionalidades basicas para o cliente e simplificando a conexão com o cliente além de permitir uma sefurança maior por conta do cliente não ter que instnaciar e manipular o objeto diretamente

```python
from abc import ABCMeta, abstractmethod
```

```python
# objecT- class abstract
class CurrentAccount(metaclass= ABCMeta):
    
    @abstractmethod
    def pay(self):
        pass
```

## vantagens

* Os proxies podem ajudar a melhorar o desempenho da aplicação ao fazer caching de objetos
pesados ou, geralmente, de objetos acessados com frequência

* Os proxies também autorizam o acesso a um objeto real; portanto este padrão auxilia na
delegação somente se as permissões estiverem corretas

* Os proxies remotos também facilitam a interação com servidores remotos, que podem
funcionar como conexões de rede e de bancos de dados, além de poderem ser usados para
monitorar sistemas.
