# Command

E um pad~rao que funciona ocmo um instalador primerio ele recebe as informações e asa acões para no final executar todas as acões de uma vez

```python
from abc import ABCMeta, abstractmethod
```

```python
# command - abstract class 
class Order (metaclass=ABCMeta):
    
    @abstractmethod
    def execute(self):
        pass
```

```python
# commmanc concrete - command A
class Buy(Order):
    
    def __init__(self, action):
        self.act = action
    
    def execute(self):
        self.act.buy()
        
# commmanc concrete - command B
class Sell(Order):
    def __init__(self,action):
        self.act = action
    
    def execute(self):
        self.act.sell()
```

```python
# receiver
class Action():
    
    def buy(self):
        print("buy products in store")
    
    def sell(sell):
        print("sell my products...")
```

```python
# invoker
class Store():
    
    def __init__(self):
        self.__cart_list = []
    
    def added_action_cart(self,action):
        self.__cart_list.append(action)
        return None
    
    def finish_cart(self):
        for index in self.__cart_list:
            index.execute()
            self.__cart_list = []
        return True
```

```python
action = Action()
buy = Buy(action)
sell = Sell(action)
```

```python
store = Store()
store.added_action_cart(buy)
store.added_action_cart(buy)
store.added_action_cart(sell)
store.added_action_cart(sell)
store.finish_cart()
```

# vantagens

* Desacopla as classes que invocam a operação do objeto

* Permite criar uma sequência de comandos oferecendo um sistema de fila

* Podemos facilmente adicionar um novo comando sem alterar o código existente
